## Seaborn Visualization

```python
import seaborn as sns     # Importing seaborn library for data visualization
```

### Distplot
It is also kown as distribution plot. It is used to plot histogram with some other variations like Kdeplot. Kdeplot is method of visualizing the distribution of data in curve.

kde - Kernal Density Estimator

```python
x = [1,2,3,4,5,6]
sns.distplot(x, color='red')
```

## Line Plot
It is mainly used to visualize the data in the form of time series.

```python
x = [1,2,3,4,5,6]
y = [5,6,3,2,1,8]
sns.lineplot(x,y)
```

## Scatter Plot
It is used to observe the relationship between variables.

```python
sns.scatterplot(x,y)
```

## Bar Plot

```python
x = ['a', 'b', 'c']
y = [1,2,3]
sns.barplot(x,y)
```

## Box Plot
It shows us the median,min,max,first quartile and third quartile. all of these can be accessed using df.describe() which gives information of min, max, first quartile and third quartile.

```python
x = [1,2,3,4]
sns.boxplot(x)
```

## Lm Plot
It shows a line representing a linear regression model with data points.

## Plotting univariate distributions
The most convenient way to take a quick look at a univariate distribution in seaborn is the distplot() function. By default this will draw a histogram and fit a KDE

```python
sns.distplot(auto['highway_mpg'])
```

## Ploting Bivariate distributions
It creates a scatterplot of the two variables along with the histogram of each next to the appropriate axes.

```python
sns.jointplot(auto['engine_size'], auto['horsepower']);
```

```python
sns.jointplot(auto['engine_size'], auto['horsepower'], kind='hex');
```

```python
sns.jointplot(auto['engine_size'], auto['horsepower'], kind='kde');
```

## Pair Plot

```python
sns.pairplot(auto)
```

```python
sns.pairplot(auto, diag_kind = 'kde')     # diag_kind means diagonal kind, diagonal graph will show KDE
```

```python
sns.pairplot(auto[['normalized_losses', 'engine_size', 'horsepower']])   # Selective columns
```

### Ploting with Categorical Data
In a strip plot, the scatterplot points will usually overlap. This makes it difficult to see the full distribution of data. One easy solution is to adjust the positions using "jitter"

```python
sns.stripplot(auto['fuel_type'], auto['horsepower'], jitter=True)
```

### Swarm Plot 
A different approach using swarmplot() function, It plots each scatterplot points on the categorical axis, and avoid overlapping.

```python
sns.swarmplot(auto['fuel_type'], auto['horsepower'])
```

### Box Plot

```python
sns.boxplot(auto['fuel_type'], auto['horsepower']);
```

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'], hue=auto['fuel_type']);
```

### Relplot
Scatterplot is used to visualize relationships between two variables. to draw the scatterplot we will be using the relplot() function of the seaborn library.

```python
sns.relplot(x='engine_size', y='horsepower', hue='make', data = auto)
```
