### Matplotlib Visualization

```python
import matplotlib.pyplot as plt     # Library required for Visualization of Data, pyplot for matplotlib library.
```

```python
x=[1,2,3,4,5,6,7,7,8,8,9]           
y=[2,3,4,5,6,7,8,8,9,9,8]
plt.plot(x,y)
plt.show()
```

```python
plt.title("Graph")       # Title of the Graph
plt.xlabel("X axis")     # Title of the X-axis
plt.ylabel("Y axis")     # Title of the Y-axis
plt.plot(x,y)            # pyplot is arranging the graph
plt.show()               # Display the Graph
```

```python
plt.plot(x,y,'ro')          # 'ro' for red circles
plt.show()
```

```python
plt.plot(x,y,'b')        # 'b' for blue line
plt.show()
```

```python
plt.plot(x,y,'-g', marker = 'o', ms = 20, mec = 'red', mfc = 'orange')     
# '-g' for green line
# marker o for O shape, s for Square Shape, d for Diamond shape
# ms is marker size
# mec is marker edge colour
# mfc is marker fill color
plt.show()
```

```python
plt.plot(x,y,'--')        # '--' for dashed line
plt.show()
```

```python
plt.plot(x,y,'^k:')        # '^k:' for black triangle up marker
plt.show()
```

```python
# representing categorical data like name, type etc.
players=["Sachin", "Yuvraj", "Dhoni"]
scores=[80, 90, 95]
plt.bar(players,scores)
plt.show()
```

```python
# some addition to the BAR plot
plt.barh(players,scores, color='orange')    # barh for horizontal bar graph
plt.title("Title of the Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
# Horizotal BAR plot
plt.bar(players,scores, color='orange')
plt.title("Title of the Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
# some addition to the Line Graph
plt.plot(players,scores, color='red', linewidth = '12.5')
plt.title("Title of the Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

## Subplot

#### Subplot is used for ploting 2 or more plots in one figure. It has three arguments,
nrows- number of rows,
ncols- number of columns,
index- number of indexes.

```python
plt.figure(figsize=(9,3))                                 # figsize is for defining the size of each figure
plt.subplot(131)                                          # (nrows,nols,index)
plt.bar(players,scores)
plt.subplot(132)
plt.scatter(players,scores)
plt.subplot(133)
plt.plot(players,scores)
plt.suptitle("Cricket Scores in Subplot")
plt.show()
```

```python
# Plotig multiple rows
plt.figure(figsize=(9,3))
plt.subplot(2,3,1)
plt.bar(players,scores)
plt.subplot(2,3,2)
plt.scatter(players,scores)
plt.subplot(2,3,3)
plt.plot(players,scores)
plt.subplot(2,3,4)
plt.plot(players,scores, color='r')
plt.subplot(2,3,5)
plt.scatter(players,scores, color='g')
plt.subplot(2,3,6)
plt.bar(players,scores, color='orange')
plt.suptitle("Cricket Scores in Subplot")
plt.show()
```

### Pie-Chart

```python
Team = ['RR','MI', 'RCB', 'KLP']
win_percent = [20,35,28,17]
plt.pie(win_percent, labels=Team)
plt.show()
```

```python
# Some addition (designing) in the Pie-Chart
plt.pie(win_percent, labels=Team, autopct='%1.2f%%', shadow=True, startangle=90)
plt.title('Score Card')
plt.legend()
plt.show()
```

```python
myexplode = [0.2, 0.5, 0.1, 0]
plt.pie(win_percent, labels=Team, explode=myexplode, autopct='%1.2f%%', shadow=True)
plt.show()
```

```python
myexplode = [0.2, 0.5, 0.1, 0]
mycolors = ["grey", "pink", "orange", "#4CAF50"]

plt.pie(win_percent, labels=Team, explode=myexplode, autopct='%1.2f%%', shadow=True, colors=mycolors)
plt.legend()
plt.show()
```

### Histogram

#### It is used to represent  group of Data into logical rages or bins.

```python
height = [132,122,145,165,145,135,137,160,122,111,190,189,178,155]
bins = [100,120,140,160,180,200]
plt.hist(height, bins, histtype = 'bar', rwidth = 0.8, color = 'purple')
plt.title('Height of Students')
plt.show()
```

### Scatter Plot

```python
x1 = [2,6,15]
y1 = [12,14,6]
plt.scatter(x1,y1)
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x1 = [2,6,15]
y1 = [12,14,6]
x2 = [12,9,13,9]
y2 = [10,11,14,6]
plt.scatter(x1,y1, color='green')
plt.scatter(x2,y2, color='red')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar()
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
height = np.array([100,110,80,87,89,95,86,59,55,94,98,78,77,68])
weight = np.array([105,125,85,85,75,86,98,85,45,85,65,75,98,58])
fig = plt.figure()
ax =plt.axes(projection='3d')
ax.scatter3D(height,weight)     # This is used to plot 3D scatter
plt.title("3D Scatter Plot")
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```
